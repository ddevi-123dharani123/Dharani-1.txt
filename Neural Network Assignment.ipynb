```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define activation functions

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
# Sample Data
x1 = 3
x2 = 2
y_actual = 17
```

```python
# Initialize random weights w1 and w2 between 1 and 10
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)
```

```python
# Learning rate
learning_rate = 0.1
```

```python
# Lists to store error and y_pred across epochs
errors = []
y_preds = []
```

```python
# Tranining (feed forward & Back Propagation) iteratively

epochs = 20

for epoch in range(epochs):
    #feed forward 
    
    y_pred = x1*w1 + x2*w2
    y_preds.append(y_pred)
    
    # Error Calculation
    
    error = (y_actual - y_pred)**2
    errors.append(error)
    
    # Back Propagation
    
    back_pro = -2*(y_actual - y_pred)
    dy_pred_dw1 = x1
    dy_pred_dw2 = x2
    
    dw1 = back_pro * dy_pred_dw1
    dw2 = back_pro * dy_pred_dw2
    
    w1 -= learning_rate * dw1
    w2 -= learning_rate * dw2
```

```python
# Visualization

plt.figure(figsize = (12,6))

plt.subplot(1,2,1)
plt.plot(range(epochs) , errors ,marker = "o")
plt.title("Error progression")
plt.xlabel("epochs")
plt.ylabel("error")


plt.subplot(1, 2, 2)
plt.plot(range(epochs), y_preds, marker='o')
plt.axhline(y=y_actual, color='g', linestyle='--', label='Actual')
plt.title("y_pred Progression")
plt.xlabel("Epoch")
plt.ylabel("y_pred")
plt.legend()

plt.show()
```
